```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from patsy import dmatrices
import json
```

```python
# load and clean data
data3 = pd.read_csv("movieRTAndBudgetData.csv")
data3 = data3.dropna()
data3.head()
data3 = data3[(data3["Worldwide"] > 0 ) & (data3["CriticsScore"] > 0)]

# Change to millions
data3["Worldwide"] = data3["Worldwide"] / 1e6
data3["Domestic"] = data3["Domestic"] / 1e6
data3["Budget"] = data3["Budget"] / 1e6


# add more columns
data3["PR"] = data3["Worldwide"] / data3["Budget"]
data3["lW"] = np.log(data3["Worldwide"])
data3["lB"] = np.log(data3["Budget"])
data3.head()
```

```python

# Plot relationship between Audience Score and Critics Score
data3.plot("AudScore","CriticsScore",kind="Scatter",alpha=0.5,title="Critics v/s Audience",figsize=(9,9))

# Plot budget and revenue
data3.plot("Budget","Worldwide",kind="Scatter",alpha=0.5,title="Revenue v/s Budget",figsize=(9,9))

# Plot Revenue and Critics Score
data3.plot("CriticsScore","Worldwide",kind="Scatter",alpha=0.5,title="Revenue v/s Critics Score",figsize=(9,9))

# Plot Revenue and Critics Score
data3.plot("AudScore","Worldwide",kind="Scatter",alpha=0.5,title="Revenue v/s Audience Score",figsize=(9,9))
```

```python

# split data by quantiles and see if there's a relationship
data3["BudgetQ"] = pd.qcut(data3["Budget"],np.linspace(0.,1.,5))
for i in pd.unique(data3["BudgetQ"]):
    data4 = data3[data3["BudgetQ"] == i]
    y, X = dmatrices("Worldwide ~ Budget + CriticsScore", data=data4, return_type="dataframe")
    model = sm.OLS(y, X)
    res = model.fit()
    print "Budget between",i
    print res.summary()
    data4.plot("CriticsScore","Worldwide",kind="Scatter",alpha=0.5,title="Revenue v/s Critics Score. Budget in:" + str(i),figsize=(9,9))
```

```python

# split data by quantiles and see if there's a relationship
data3["BudgetQ"] = pd.qcut(data3["Budget"],np.linspace(0.,1.,5))
for i in pd.unique(data3["BudgetQ"]):
    data4 = data3[data3["BudgetQ"] == i]
    y, X = dmatrices("PR ~ Budget + CriticsScore + AudScore", data=data4, return_type="dataframe")
    model = sm.OLS(y, X)
    res = model.fit()
    print "Budget between",i
    print res.summary()
    data4.plot("CriticsScore","PR",kind="Scatter",alpha=0.5,title="Revenue/Budget v/s Critics Score. Budget in:" + str(i),figsize=(9,9))
```

```python
actors = {}
numInclude = 25

def getActors(row):
    cast = json.loads(row.Cast)
    for person in cast:
        if person["name"] in actors.keys(): actors[person["name"]] += 1
        else: actors[person["name"]] = 1
    return None

def checkActors(row,actorName):
    cast = json.loads(row.Cast)
    for person in cast:
        if person["name"] == actorName: return True
    
    return False

# get numbers of movies each actor has been in
success = data3.apply(getActors,1)

# get sorted list of actors by number of films
actorsList = sorted(actors.keys(),key=lambda x: actors[x],reverse=True)

# create binary variables for the top x actors
data4 = data3.copy()
actorsToUse = actorsList[:numInclude]
for a in actorsToUse:
    data4[a] = data4.apply(checkActors,1,args=(a,))

# regression using top actors
indepVariables = actorsToUse[:]
indepVariables.append("Budget")
indepVariables.append("CriticsScore")
indepVariables.append("AudScore")
y,X = data4["PR"],data4[indepVariables]
X = sm.add_constant(X,prepend=False)
model = sm.OLS(y, X)
res = model.fit()
print res.summary()
```
