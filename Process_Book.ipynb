```python
# Function to make dictionary of ("word", True) for Naive bayes classifier
def word_feats(words):
    return dict([(word, True) for word in words])

# Function to make dictionary of (("word", 'word'), True) for Naive bayes classifier
def bigram_word_feats(words, score_fn=BigramAssocMeasures.chi_sq, n=200):
    bigram_finder = BigramCollocationFinder.from_words(words)
    bigrams = bigram_finder.nbest(score_fn, n)
    return dict([(ngram, True) for ngram in itertools.chain(words, bigrams)])

# Function that returns dictionary of stop words from the nltk corpus
def stopword_filtered_word_feats(words):
    return dict([(word, True) for word in words if word not in stopset])
```

```python
# Function returns a trained NB Classifier
def evaluate_classifier(featx):
    
    negids = movie_reviews.fileids('neg')
    posids = movie_reviews.fileids('pos')
 
    negfeats = [(featx(movie_reviews.words(fileids=[f])), 'neg') for f in negids]
    posfeats = [(featx(movie_reviews.words(fileids=[f])), 'pos') for f in posids]
 
    negcutoff = len(negfeats)*3/4
    poscutoff = len(posfeats)*3/4
 
    trainfeats = negfeats[:negcutoff] + posfeats[:poscutoff]
    testfeats = negfeats[negcutoff:] + posfeats[poscutoff:]
 
    classifier = NaiveBayesClassifier.train(trainfeats)
    refsets = collections.defaultdict(set)
    testsets = collections.defaultdict(set)
 
    for i, (feats, label) in enumerate(testfeats):
            refsets[label].add(i)
            observed = classifier.classify(feats)
            testsets[observed].add(i)
 
    print 'accuracy:', nltk.classify.util.accuracy(classifier, testfeats)
    print 'pos precision:', nltk.metrics.precision(refsets['pos'], testsets['pos'])
    print 'pos recall:', nltk.metrics.recall(refsets['pos'], testsets['pos'])
    print 'neg precision:', nltk.metrics.precision(refsets['neg'], testsets['neg'])
    print 'neg recall:', nltk.metrics.recall(refsets['neg'], testsets['neg'])
    classifier.show_most_informative_features()
    top15_important = classifier.most_informative_features()[:15]

    return classifier
```

```python
"""
Function
--------
review_finder

Function returns a movie review dataframe for individual moviews

Parameters
----------
name : string
    The name of a movie

Returns
-------
movie_DF : Dataframe
    Dataframe with 1 column of RT critic reviews for individual movie
"""
def review_finder(name):
    try: 
        movie = str(name)
        temp_reviewDF = movie_reviewDF.set_index('Name')
        reviewSTR = json.loads(temp_reviewDF.loc[movie].Reviews_JSON)
        
        if reviewSTR == 'NA':
            #print 'Error: No movie review present'
            Movie_DF = 0
        else:
            Movie_DF = pd.DataFrame(reviewSTR, columns = [movie])

        return Movie_DF
    
    except Exception, e:
        print 'Failed in movie dataframe builder function'
        print str(e)
```

The below function uses the Naive Bayes Classifier to predict the probablility of a word being positive or negative. It then assigns a rating based on the amount of positive, negative, and netural (positive probability = negative probabilty) words in the sentence. 

We decided to review key parts of the sentence such as: 
- Verbs, which includes: VB (base form ie. take), VBD(past tense ie. took), VBG (gerund/present participle ie. taking), VBN (past participle ie. taken), VBP (singular present, non-3d ie. take), VBZ (3rd person sing. present ie. takes)
- Adjectives, which includes: JJ (base form ie. big), JJR (comparative ie. bigger), JJS (superlative ie. biggest)
- Nouns, which includes: NN (singular or mass ie. door), NNS (plural ie. doors), NNP (proper noun singular ie. John), NNPS (proper noun plural ie. Patriots)
- Adverb, which includes: RB (base ie. however, usually, etc.), RBR (comparative ie. better), RBS (superlative ie. best) 

```python
"""
Function
--------
review_sentiment

Function returns a sentiment rating for a critic's movie review

Parameters
----------
sentence: string
    The movie review

Returns
-------
review : string
    Returns neg, pos, net based on the sentiment analysis
"""
def review_sentiment(sentence):
    try:
        tokenized = nltk.word_tokenize(sentence)
        tagged = nltk.pos_tag(tokenized)
        stripped = re.findall(r'\(\u\'(\w*)\',\s\'(VB\w?|JJ\w?|NN\w?|RB\w?)\'', str(tagged))
        tags = re.findall(r'\(\'(\w*)\'', str(stripped))

        probs = map(lambda x: class_naive.prob_classify({x: True}), tags)
        pos_prob = zip(tags, map(lambda x: x.prob("pos"), probs))
        neg_prob = zip(tags, map(lambda y: y.prob("neg"), probs))

        pos_word = []
        neg_word = []
        for x in pos_prob:
            if (x[1] > 0.5) == True: pos_word.append(x)
            elif (x[1] < 0.5) == True: neg_word.append(x) 

        if len(pos_word) > len(neg_word): review = 'fresh' 
        elif len(pos_word) < len(neg_word): review = 'rotten'
        else: review = 'net'

        return review
        
    except Exception, e:
        print 'Failed in sentiment review function'
        print str(e)
```

This function is similiar to the above function, however it uses the Naive Bayes Classifier trained using bigrams. Bigrams are two adjencent words from a sentence/s that are grouped together.

```python
"""
Function
--------
review_bigram

Function returns a sentiment rating for a critic's movie review

Parameters
----------
sentence: string
    The movie review

Returns
-------
score : int
    Overall score of the movie based on the critic's reviews
flag: str
    Either "Positive", Negative"
"""

def review_bigram(sentence):
        bigram_finder = BigramCollocationFinder.from_words(nltk.word_tokenize(sentence))
        bigram = bigram_finder.nbest(BigramAssocMeasures.pmi, 70)
        
        probs = map(lambda x: class_bigram.prob_classify({x: True}), bigram)
        pos_prob = zip(bigram, map(lambda x: x.prob("pos"), probs))
        neg_prob = zip(bigram, map(lambda y: y.prob("neg"), probs))
        
        pos_word = []
        neg_word = []
        for x in pos_prob:
            if (x[1] > 0.5) == True: pos_word.append(x)
            elif (x[1] < 0.5) == True: neg_word.append(x)
        
        if len(pos_word) > len(neg_word): review = 'fresh' 
        elif len(pos_word) < len(neg_word): review = 'rotten' 
        else: review = 'net'
        
        return review  
```

```python
"""
Function
--------
assign_reviews

Function that assigns overall score to a movie by reviewing all critic's reviews
for a particular film. Similiar to the Rotten Tomatoe freshness score

Parameters
----------
data: DF
    data is moive's dataframe of critic revies 
classifer: str
    Either 'naive' or 'bigram' 

Returns
-------
score : int
    Percentage of positive reviews over all critic reviews
flag : string
    Fresh, Rotten, or Netural
dataDF : Dataframe
    Dataframe of all critic reviews with the sentiment analysis score attached
"""

def assign_reviews(dataDF, classifier):
    try:
        if len(dataDF) <= 2:
            return 0, 0, 0
            raise ValueError("There are not enough reviews for this movie")

        sentiment_reviews_score = []
        if classifier == 'naive':
            for index, row in dataDF.iterrows():
                    sentiment_reviews_score.append(review_sentiment(row[0]))
        elif classifier == 'bigram':
            for index, row in dataDF.iterrows():
                    sentiment_reviews_score.append(review_bigram(row[0]))

        net_count = list(sentiment_reviews_score).count('net')
        fresh_count = list(sentiment_reviews_score).count('fresh')
        rotten_count = list(sentiment_reviews_score).count('rotten')

        all_reviews = len(dataDF)-net_count
        freshness_score = np.round((fresh_count*1.0/all_reviews*1.0)*100)
        
        return sentiment_reviews_score, freshness_score, fresh_count
    
    except Exception, e:
        print 'Failed in assign_reviews function'
        print str(e)
```

```python
def compare_results(row):
    try:
        if row.Freshness == row.Sentiment: compare = True
        else: compare = False
        return compare
    
    except Exception, e:
        print 'Failed in compare_results function'
        print str(e)
```

```python
# Given a list of movies and a classifier, function classifies all movies and returns a percentage and rating
def rate_movies(movie_names, classifier):
    try:
        classifier_score = []
        all_movies_fresh_score = []
        for x in movie_names:
            print x
            temp_movieDF = review_finder(x)
            sentiment_reviews, movies_fresh_score, fresh_reviews = assign_reviews(temp_movieDF, classifier)
            temp_movieDF['Sentiment'] = sentiment_reviews
            
            if len(temp_movieDF) > 2:
                temp_movieDF['Freshness'] = total_budget_reviewDF.loc[x].Freshness.values
                temp_movieDF['Compare'] = temp_movieDF.apply(compare_results,1)
                classifier_score.append(np.round((temp_movieDF.Compare.sum()*1.0)/(len(sentiment_reviews)*1.0)*100.0))
                all_movies_fresh_score.append(movies_fresh_score)
            
            else: 
                classifier_score.append(0)
                all_movies_fresh_score.append(0)
                
        return all_movies_fresh_score, classifier_score
    
    except Exception, e:
        print 'Failed in rate_movie function'
        print str(e)
```

Overall, the Naive Bayes with individual words as well as the Bigram approach don't predict the sentiment of a critic's review very well. So we looked closer at the sentiment scores of both approaches to find patterns in the data. After further analysis, we noticed that the Naive Bayes with individual words tends to do a better job of predicting postivie critic sentiment while it srtuggles to predict negative reviews. This makes sense when looking at the positive recall score verus the negative recall score. The Bigram Naive Bayes apporach does a better job of predicting negative reviews, as indicated by it's higher negative recall score.

So we decided to use an combination of both approaches to predict the critic's freshness score. Individual NB for the critic's reviews that are higher than 50% and the Bigram approach for the review's that are less than 50%.

```python
def get_best(row):
    try:
        if row.CriticsScore < 50.0: return row['Sentiment_Bi_Score']
        elif row.CriticsScore > 50.0: return row['Sentiment_Score']
    
    except Exception, e:
        print 'Failed in get_best function'
        print str(e) 
```
