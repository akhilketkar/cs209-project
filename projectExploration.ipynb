```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("movieBudgets.csv",encoding="utf8")
```

```python
data["PR"] = data["Worldwide"] / data["Budget"]
data.head()
```

```python
data2 = data[ (data["Budget"] > 0) & (data["Worldwide"] > 0) ]

# data2.plot("Budget","Worldwide",kind="Scatter",alpha=0.5,xlim=(-0.25e8,3.e8),ylim=(-0.25e9,2e9),figsize=(16,9))
slope, intercept, r_value, p_value, std_err = stats.linregress(data2.Budget,data2.Worldwide)
# plt.plot([0,4.e8],[intercept,intercept+slope*4.e8])
data2.plot("Budget","PR",kind="Scatter",alpha=0.5,figsize=(16,9),ylim=(0,1.))
```
