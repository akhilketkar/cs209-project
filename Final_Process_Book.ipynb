```python
actors = {}
numInclude = 25

def getActors(row):
    cast = json.loads(row.Cast)
    for person in cast:
        if person["name"] in actors.keys(): actors[person["name"]] += row.Worldwide
        else: actors[person["name"]] = row.Worldwide
    return None

def checkActors(row,actorName):
    cast = json.loads(row.Cast)
    for person in cast:
        if person["name"] == actorName: return True
    
    return False

# get numbers of movies each actor has been in
success = dataMain.apply(getActors,1)

# get sorted list of actors by number of films
actorsList = sorted(actors.keys(),key=lambda x: actors[x],reverse=True)

# create binary variables for the top x actors
dataMainActors = dataMain.copy()
actorsToUse = actorsList[:numInclude]
actorsToUse.append("Tom Cruise")
for a in actorsToUse:
    dataMainActors[a] = dataMainActors.apply(checkActors,1,args=(a,))

# regression using top actors
indepVariables = actorsToUse[:]
indepVariables.append("logBudget")
indepVariables.append("CriticsScore")
# indepVariables.append("AudScore")

for i in pd.unique(dataMain["BudgetQ"]):
    dataPRQ = dataMainActors[dataMainActors["BudgetQ"] == i]
    y,X = dataPRQ["logWorldwide"],dataPRQ[indepVariables]
    model = sm.OLS(y, X)
    res = model.fit()
    print "Budget between",i
    print res.summary()
```

## For the second piece of our project we used Natural Language Processing to assess movie critic reviews. We sought out to compare the language that they use in the review against their actual quantative score of the movie.

```python
"""
This code is referenced from
http://streamhacker.com/2010/05/24/text-classification-sentiment-analysis-stopwords-collocations/
I found the tutorials to be verify informative and gave code examples for 
how to train the Naive Bayes Classifier with a bag of words
"""
# Function to make dictionary of ("word", True) for Naive bayes classifier
def word_feats(words):
    return dict([(word, True) for word in words])

# Function to make dictionary of (("word", 'word'), True) for Naive bayes classifier
def bigram_word_feats(words, score_fn=BigramAssocMeasures.chi_sq, n=200):
    bigram_finder = BigramCollocationFinder.from_words(words)
    bigrams = bigram_finder.nbest(score_fn, n)
    return dict([(ngram, True) for ngram in itertools.chain(words, bigrams)])
```

This function below code trains the Naive Bayes Classifier on the NLTK “movie reviews” corpus. It uses the bag of words approach by splitting the NLTK corpus named "movie reviews" in to two different labels - “pos” and “neg”. The words from the positive and negative reviews are then split in to training and test splits by taking 75% of the total words for training and 25% for test.

The Naive Bayes Classifier treats features independently by determining which feature is the most common occuring within a label. These features then “vote” against one another to determine the label of a feature. The Naive Bayes then returns the probability that a feature belongs to each respective label. 

```python
"""
This code is referenced from
http://streamhacker.com/2010/05/24/text-classification-sentiment-analysis-stopwords-collocations/
"""
# Function returns a trained NB Classifier
def evaluate_classifier(featx):
    
    negids = movie_reviews.fileids('neg')
    posids = movie_reviews.fileids('pos')
 
    negfeats = [(featx(movie_reviews.words(fileids=[f])), 'neg') for f in negids]
    posfeats = [(featx(movie_reviews.words(fileids=[f])), 'pos') for f in posids]
 
    negcutoff = len(negfeats)*3/4
    poscutoff = len(posfeats)*3/4
 
    trainfeats = negfeats[:negcutoff] + posfeats[:poscutoff]
    testfeats = negfeats[negcutoff:] + posfeats[poscutoff:]
 
    classifier = NaiveBayesClassifier.train(trainfeats)
    refsets = collections.defaultdict(set)
    testsets = collections.defaultdict(set)
 
    for i, (feats, label) in enumerate(testfeats):
            refsets[label].add(i)
            observed = classifier.classify(feats)
            testsets[observed].add(i)
 
    print 'accuracy:', nltk.classify.util.accuracy(classifier, testfeats)
    print 'pos precision:', nltk.metrics.precision(refsets['pos'], testsets['pos'])
    print 'pos recall:', nltk.metrics.recall(refsets['pos'], testsets['pos'])
    print 'neg precision:', nltk.metrics.precision(refsets['neg'], testsets['neg'])
    print 'neg recall:', nltk.metrics.recall(refsets['neg'], testsets['neg'])
    
    classifier.show_most_informative_features()
    top15_important = classifier.most_informative_features()[:15]
    
    return classifier
```

The Naive Bayes Classifier is approximately 73% accurate when trained with individual words from the pos/neg bag of words. The classifier has very good positive recall (98%), which means that it doesn’t predicate many false negatives (predicting a negative word when it is actually positive).  However, the positive precision is somewhat lower (~65%). This means that classifier incorrectly predicts positives often leading to false positives (predicts a positive word when the word is actually negative). 

Conversely, negative precision is very high (~96 %), which means that any word that is classified as negative is very likely to be correct (low false positives). The negative recall is very low (~48%), which means that a high number of words are incorrectly classified as negative (high false negatives).

The function also outputs the most important features with their respective classes. These are the most occurring features within the positive and negative classes. Most of the words are adjectives which makes sense as they are descriptive words.

```python
"""
Function
--------
review_finder

Function returns a movie review dataframe for individual moviews

Parameters
----------
name : string
    The name of a movie

Returns
-------
movie_DF : Dataframe
    Dataframe with 1 column of RT critic reviews for individual movie
"""
def review_finder(name):
    try: 
        movie = str(name)
        temp_reviewDF = movie_reviewDF.set_index('Name')
        reviewSTR = json.loads(temp_reviewDF.loc[movie].Reviews_JSON)
        
        if reviewSTR == 'NA':
            Movie_DF = 0
        else:
            Movie_DF = pd.DataFrame(reviewSTR, columns = [movie])

        return Movie_DF
    
    except Exception, e:
        print 'Failed in movie dataframe builder function'
        print str(e)
```

The below function uses the Naive Bayes Classifier to predict the probablility of a word being positive or negative. It then assigns a rating based on the amount of positive, negative, and netural (positive probability = negative probabilty) words in the sentence. 

We decided to review key parts of the sentence such as: 
- Verbs, which includes: VB (base form ie. take), VBD (past tense ie. took), VBG (gerund/present participle ie. taking), VBN (past participle ie. taken), VBP (singular present, non-3d ie. take), VBZ (3rd person sing. present ie. takes)
- Adjectives, which includes: JJ (base form ie. big), JJR (comparative ie. bigger), JJS (superlative ie. biggest)
- Nouns, which includes: NN (singular or mass ie. door), NNS (plural ie. doors), NNP (proper noun singular ie. John), NNPS (proper noun plural ie. Patriots)
- Adverb, which includes: RB (base ie. however, usually, etc.), RBR (comparative ie. better), RBS (superlative ie. best) 

```python
"""
Function
--------
review_sentiment

Function returns a sentiment rating for a critic's movie review

Parameters
----------
sentence: string
    The movie review

Returns
-------
review : string
    Returns neg, pos, net based on the sentiment analysis
"""
def review_sentiment(sentence):
    try:
        tokenized = nltk.word_tokenize(sentence)
        tagged = nltk.pos_tag(tokenized)
        stripped = re.findall(r'\(\u\'(\w*)\',\s\'(VB\w?|JJ\w?|NN\w?|RB\w?)\'', str(tagged))
        tags = re.findall(r'\(\'(\w*)\'', str(stripped))

        probs = map(lambda x: class_naive.prob_classify({x: True}), tags)
        pos_prob = zip(tags, map(lambda x: x.prob("pos"), probs))
        neg_prob = zip(tags, map(lambda y: y.prob("neg"), probs))

        pos_word = []
        neg_word = []
        for x in pos_prob:
            if (x[1] > 0.5) == True: pos_word.append(x)
            elif (x[1] < 0.5) == True: neg_word.append(x) 
        
        if len(pos_word) > len(neg_word): review = 'fresh' 
        elif len(pos_word) < len(neg_word): review = 'rotten'
        else: review = 'net'
        
        return review
        
    except Exception, e:
        print 'Failed in sentiment review function'
        print str(e)
```

This function is similiar to the above function, however it uses the Naive Bayes Classifier trained using bigrams. Bigrams are two adjencent words from a sentence/s that are grouped together.

```python
"""
Function
--------
review_bigram

Function returns a sentiment rating for a critic's movie review

Parameters
----------
sentence: string
    The movie review

Returns
-------
Rating : str
    Either "Positive", Negative"s
"""

def review_bigram(sentence):
    try:
        bigram_finder = BigramCollocationFinder.from_words(nltk.word_tokenize(sentence))
        bigram = bigram_finder.nbest(BigramAssocMeasures.chi_sq, 80)
        bigram_dict = dict([(ngram, True) for ngram in itertools.chain(bigram)])

        prob_bigram = class_bigram.prob_classify(bigram_dict)
        pos_prob = prob_bigram.prob("pos")
        neg_prob = prob_bigram.prob("neg")

        #print "pos:", pos_prob, "neg:", neg_prob, '\n'

        if pos_prob > neg_prob:
            return 'fresh'
        elif neg_prob > pos_prob:
            return 'rotten'
    
    except Exception, e:
        print 'Failed in review_bigram function'
        print str(e)
```

```python
"""
Function
--------
assign_reviews

Function that assigns overall score to a movie by reviewing all critic's reviews
for a particular film. Similiar to the Rotten Tomatoe freshness score

Parameters
----------
data: DF
    data is moive's dataframe of critic revies 
classifer: str
    Either 'naive' or 'bigram' 

Returns
-------
sentiment_reviews_score : int
    Percentage of positive reviews over all critic reviews
freshness_score : int
    The amount of times the classifier prediction matches the critics review
fresh_count : int
    The total amount of reviews that are predicted as fresh
"""

def assign_reviews(dataDF, classifier):
    try:
        if len(dataDF) <= 2:
            return 'NA', 'NA', 'NA'
            raise ValueError("There are not enough reviews for this movie")

        sentiment_reviews_score = []
        if classifier == 'naive':
            for index, row in dataDF.iterrows():
                    sentiment_reviews_score.append(review_sentiment(row[0]))
        elif classifier == 'bigram':
            for index, row in dataDF.iterrows():
                    sentiment_reviews_score.append(review_bigram(row[0]))

        net_count = list(sentiment_reviews_score).count('net')
        fresh_count = list(sentiment_reviews_score).count('fresh')
        rotten_count = list(sentiment_reviews_score).count('rotten')

        all_reviews = len(dataDF)-net_count
        freshness_score = np.round((fresh_count*1.0/all_reviews*1.0)*100)
        
        return sentiment_reviews_score, freshness_score, fresh_count
    
    except Exception, e:
        print 'Failed in assign_reviews function'
        print str(e)
```

```python
# Function that compares the Freshness score to the sentiment score for each movie
def compare_results(row):
    try:
        if row.Freshness == row.Sentiment: compare = True
        else: compare = False
        return compare
    
    except Exception, e:
        print 'Failed in compare_results function'
        print str(e)
```

```python
"""
Function
--------
assign_reviews

Given a list of movies and a classifier, function classifies all movies and returns a sentiment percentage 
and the classifier score

Parameters
----------
movie_names: list
    list of movies to review
classifer: str
    Either 'naive' or 'bigram' 

Returns
-------
all_movies_fresh_score : list of ints
    Sentiment percent score, the ratio of "fresh" reviews to all reviews for each movie
classifier_score : list of ints
    Classifier score, how well the classifier predicted the critic review for each movie
"""
def rate_movies(movie_names, classifier):
    try:
        classifier_score = []
        all_movies_fresh_score = []
    
        for x in movie_names:
            temp_movieDF = review_finder(x)
            print '###########', x, '###########'
            
            sentiment_reviews, movies_fresh_score, fresh_reviews = assign_reviews(temp_movieDF, classifier)
            temp_movieDF['Sentiment'] = sentiment_reviews
            temp_movieDF['Freshness'] = total_budget_reviewDF.loc[x].Freshness.values
            temp_movieDF['Compare'] = temp_movieDF.apply(compare_results,1)
            
            classifier_score.append(np.round((temp_movieDF.Compare.sum()*1.0)/(len(sentiment_reviews)*1.0)*100.0))
            all_movies_fresh_score.append(movies_fresh_score)
            print "Fresh Reviews:", fresh_reviews, "Total Reviews:", len(temp_movieDF), "Score:", movies_fresh_score, '\n'

        return all_movies_fresh_score, classifier_score
    
    except Exception, e:
        print 'Failed in rate_movie function'
        print str(e)
```
